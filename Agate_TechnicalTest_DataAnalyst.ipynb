Which of the following states contributed approximately 25% of the combined production of all the 5 country in the year 2002?

a. Indonesia

b. Singapore

c. Malaysia

d. Thailand

e. Philippines

```
# data in year 2002
year_2002 = {"Indonesia":22,"Singapore":19,"Malaysia":26,"Thailand":11,"Philippines":13}
total_production_2002 = sum(year_2002.values())
print("in year 2002:")
near_25percent = 999
answer = ""
for country in year_2002.keys():
  contribution = round((year_2002[country]/total_production_2002*100),2)
  print(country,"contributed:",contribution,"%.")
  if abs(25 - contribution) < near_25percent:
    near_25percent = abs(25 - contribution)
    answer = country
print("")
print("So, the answer is",answer.upper())

```

Which of the following statements is/are true regarding the consumption of chocobar?

a. The percentage change in consumption of chocobar over the previous year is the same every year.

b. The rate of fall of consumption chocobar is increasing steadily.

c. The steepest fall in the consumption of chocobar follows the steepest increase in consumption.

d. The consumption is falling and increasing in alternate years.

e. All the statements are true.


```
# data in consumption chocobar (in '000 bars)
year = [1993,1994,1995,1996,1997,1998,1999]
bars = [124,118,128,92,134,126,122]
i = 0
while i < len(year)-1:
  change = bars[i+1] - bars[i]
  percent_change = round((change/bars[i]),2)
  print("From",year[i],"to",year[i+1],", change of bars consumption:",change,"or",percent_change,"%.")
  i += 1
print("")
print("Therefore, the answer is:")
print("The steepest fall in the consumption of chocobar follows the steepest increase in consumption.")
```
