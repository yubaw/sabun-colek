Which of the following states contributed approximately 25% of the combined production of all the 5 country in the year 2002?

````
# data in year 2002
year_2002 = {"Indonesia":22,"Singapore":19,"Malaysia":26,"Thailand":11,"Philippines":13}
total_production_2002 = sum(year_2002.values())
print("in year 2002:")
near_25percent = 999
answer = ""
for country in year_2002.keys():
  contribution = round((year_2002[country]/total_production_2002*100),2)
  print(country,"contributed:",contribution,"%.")
  if abs(25 - contribution) < near_25percent:
    near_25percent = abs(25 - contribution)
    answer = country
print("")
print("So, the answer is",answer.upper())

````

5. Over the period 1993-94 to 1997-98, there has been a/an _____ in fertiliser subsidy.

image.png

__a. Increase__ (ANSWER)

b. Decrease

c. Constant Expenditure

d. Cannot Say

e. Inconsistent Expenditure

````
# fertiliser subsidy data (in million) from 93-94 to 97-98
year = [93.94,94.95,95.96,96.97,97.98]
indigenous_fertiliser = [3000,3400,3350,3300,4800]
imported_fertiliser = [200,1142,1039,1499,1000]
i = 0
change_indigenous = []
change_imported = []
while i < len(year)-1:
  change_indigenous.append(indigenous_fertiliser[i+1]-indigenous_fertiliser[i])
  change_imported.append(imported_fertiliser[i+1]-imported_fertiliser[i])
  print("Change of indigenous fertiliser subsidy",year[i],"to",year[i+1],"is:",
        indigenous_fertiliser[i+1]-indigenous_fertiliser[i])
  print("Change of imported fertiliser subsidy",year[i],"to",year[i+1],"is:",
        imported_fertiliser[i+1]-imported_fertiliser[i])
  i += 1
total_change_indigenous = sum(change_indigenous)
total_change_imported = sum(change_imported)
print("----------------------------------")
print("Total change in indigenous fertiliser subsidy:", total_change_indigenous)
print("Total change in imported fertiliser subsidy:", total_change_imported)
print(" ")
if total_change_imported + total_change_indigenous > 0:
  print("Conclusion: There has been an INCREASE in fertiliser subsidy")
elif total_change_imported + total_change_indigenous < 0:
  print("Conclusion: There has been a DECREASE in fertiliser subsidy")
else:
  print("Conclusion: There has been a CONSTANT expenditure in fertiliser subsidy")
````

1. Lets say you have two SQL tables: authors and books. The authors dataset has 1M+ rows heres the first six rows:

image.png

The books dataset also has 1M+ rows and heres the first six:

image.png

Create an SQL query that shows the TOP 3 authors who sold the less books in total!


__Solution__ 

The relation between _authors datasheet_ and _books datasheet_ is in book_name column. So, we need to

*   combine two datasheet with INNER JOIN based ON similarity in book_name
*   do GROUP BY author_name to get value of total sold_copies (using SUM)
*   do ORDER BY total sold_copies in ascending order (from smallest to largest)
*   finally, use LIMIT to show author_name with 3 smallest value in total sold_copies

```
SELECT authors_datasheet.author_name, 
SUM(books_datasheet.sold_copies) AS total_sold_copies
FROM authors_datasheet
INNER JOIN books_datasheet
ON authors_datasheet.book_name = books_datasheet.book_name
GROUP BY authors_datasheet.author_name
ORDER BY total_sold_copies ASC LIMIT 3;
```

Note: If the event_date_time columns format doesnt look familiar, google epoch timestamp!

Write an SQL query to find out how many users inserted more than 2000 but less than 4000 images in their presentations!

__Solution__

Epoch timestamp can represent all time zones at once, therefore it must me unique for each user_id. So, the repetition specific user ID in user_id column also represent repetition of how many image inserted. 

For solving the problem, we need to

*   SELECT user_id and COUNT each user ID repetition

3. You have two SQL tables! The first one is called employees and it contains the employee names, the unique employee ids and the department names of a company.

image.png

The second one is named salaries. It holds the same employee names and the same
employee ids  and the salaries for each employee.

image.png

The company has 648 employees, so both tables have 648 rows.

Print every department where the average salary per employee is over than $500!


__Solution__

using similarity ON employee_id column, we can combine _employees_ table INNER JOIN with _salaries_ table. Then we do GROUP BY _department_name_ and calculate average value of _salary_. Finally, we filter average salary bigger than 500 with HAVING.

```
SELECT employees.department_name,
(AVG(salaries.salary)) AS average_salary_per_employee
FROM employees
INNER JOIN salaries
ON employees.employee_id = salaries.employee_id
GROUP BY employees.department_name
HAVING AVG(salaries.salary) > 500
```

image.png

